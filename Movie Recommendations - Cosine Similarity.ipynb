```python
file_path = r"C:\Users\User\Downloads\archive\tmdb_5000_movies.csv" #import the 'movies' dataset
movies = pd.read_csv(file_path)
movies.head(2) #display the first two rows of the dataset
```

```python
file_path = r"C:\Users\User\Downloads\archive\tmdb_5000_credits.csv" #import the 'credits' dataset
credits = pd.read_csv(file_path)
credits.head(2) #display the first 2 rows
```

```python
movies = movies.merge(credits,on='title') #merge the 'movies' and 'credits' datasets
movies.head(2) 
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] #select needed columns, get rid of non-relevant columns
movies.head(2)
```

```python
movies.isnull().sum() #check for null values
```

```python
movies.dropna(inplace=True) #drop null values
```

```python
import ast

#changing columns from stings to lists
def convert(obj): 
    L = [] #initaiate an empty list
    for i in ast.literal_eval(obj): #loop through string
        L.append(i['name']) #append 'name' values to the list
    return L #return the list of 'name' values
```

```python
movies["genres"] = movies['genres'].apply(convert) #apply the function to the 'genres' column
movies.head(2)
```

```python
movies["keywords"] = movies['keywords'].apply(convert) #apply the function to the 'keywords' column
movies.head(2)
```

```python
# changing from strings to lists with first 3 names
def convert3(obj): 
    L = [] #intialize list
    counter = 0 #intialize a counter
    for i in ast.literal_eval(obj):
        if counter !=3: 
            L.append(i['name']) #append name
            counter += 1 #add to counter
        else:
            break #break when the counter hits 3
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3) #apply the function to the 'cast' column
movies.head(2)
```

```python
#create a function to find the director
def fetch_director(obj):
    L = [] #create a list to store names
    #append 'name' to list if the 'job' is 'director'
    for i in ast.literal_eval(obj):
        if i ['job'] == "Director": 
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director) #apply the function to the 'crew' column
movies.head(2)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) #convert 'overview' to a list and split at whitespaces
movies.head(2)
```

```python
#remove spaces 'genres', 'keywords' 'cast' and'crew' columns
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies.head(2)
```

```python
#combine 'overview', 'genres', 'keywords', 'cast', and 'crew' columns to the 'tags' column
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
movies.head(2)
```

```python
new_df = movies[["movie_id","title","tags"]] #create a new dataframe with the 'tags' column 
new_df.head(2)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')

#remove stop words from the dataset
def remove_stopwords(word_list):
    stop_words = set(stopwords.words('english'))
    filtered_words = [word.lower() for word in word_list if word.lower() not in stop_words]
    return filtered_words

new_df['tags'] = new_df['tags'].apply(remove_stopwords)
```

```python
import string #remove punctuation marks and numbers
def remove_punctuation_and_numbers(word_list):
    translator = str.maketrans(string.punctuation + string.digits, 
                               ' ' * (len(string.punctuation) + len(string.digits)))
    return [word.translate(translator) for word in word_list]


new_df['tags'] = new_df['tags'].apply(remove_punctuation_and_numbers)
```

```python
#lemmatize the 'tags' column
nltk.download('omw-1.4')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_list(tag_list):
    return [lemmatizer.lemmatize(word) for word in tag_list]

new_df['tags'] = new_df['tags'].apply(lemmatize_list)
new_df.head(2)
```

```python
#stem the 'tags' column
from nltk.stem import PorterStemmer
ps = PorterStemmer()

def stem_list(tag_list):
    return [ps.stem(word) for word in tag_list]

new_df['tags'] = new_df['tags'].apply(stem_list)
new_df.head(2)
```

```python
new_df['tags'] = new_df["tags"].apply(lambda x:" ".join(x)) #convert 'tags' back to a string
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join(x.split())) #make sure theres only one space between each word
new_df.head(2)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower()) #lowercase the 'tags' column
new_df.head(2)
```

```python
# Remove the word "get" (case insensitive)
new_df['tags'] = new_df['tags'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() != 'get']))
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(max_features=5000) #convert into a matrix
#rows = movies, term = terms
#limit terms to top 5000
#remove stopwords
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
similarity = cosine_similarity(vector) #compute similarity scores between movies and terms
```

```python
def recommend(movie): #create a function to recommend movies
    index = new_df[new_df['title'] == movie].index[0] 
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1]) #find movies with highest similarity scores
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title) #print the titles of the top 5 similar movies
```

```python
recommend('Elf') #test on "Elf"
```

```python
recommend('Spectre') #test of "Spectre"
```

```python
recommend('Toy Story') #test on "Toy Story"
```

```python
recommend("Cars") #test on "Cars"
```

```python
def find_common_tags(movie1, movie2): #create a function to display the common tags
    index1 = new_df[new_df['title'] == movie1].index[0] #find the index of the movies in the dataframe
    index2 = new_df[new_df['title'] == movie2].index[0]

    tags_movie1 = set(new_df.iloc[index1]['tags'].split(' ')) #extract tags for each movie and convert them to sets
    tags_movie2 = set(new_df.iloc[index2]['tags'].split(' '))
    
    common_tags = tags_movie1.intersection(tags_movie2) #find the common tags between the two sets
    return common_tags
```

```python
find_common_tags('Toy Story', "Small Soldiers") #find common tags between "Toy Story" and "Small Soldiers"
```

```python
find_common_tags('Toy Story', 'The Lego Movie') #find common tags between "Toy Story" and "The Lego Movie"
```
